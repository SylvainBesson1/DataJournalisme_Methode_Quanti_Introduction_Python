## Récapitulatif général des commandes de base

|                     **Commande**                    |                          **Résultat**                     |
|:---------------------------------------------------:|:---------------------------------------------------------:|
| print()                                             | Afficher un message                                       |
| nom_variable = 2024                                 | Assigner 2024 à la variable nom_variable                  |
| type()                                              | Donner le type d’une variable                             |
| str() ; int() ; float() ; bool()                    | Convertir le type d’une donnée                            |
| + ; - ; * ; / ; // ; % ; ** ; < ; <= ; == ; >= ; >  | Opérateurs arithmétiques et logiques                      |
| # …                                                 | Commentaires                                            |
| nom_variable = []                                   | Créer une liste                                           |
| .append(n)                                          | Ajouter n à la fin d'une liste                            |
| del nom_variable [n]                              | Retirer l’élément n d’une liste                           |
| len()                                               | Donner la longueur d’une variable                         |
| nom_variable = {}                                   | Créer un dictionnaire                                     |
| nom_variable ["clé"] = "valeur"                   | Ajouter une valeur et une clé à un dictionnaire           |
| .keys()  ; .values() ; .items()                     | Récupérer les clés ; les valeurs ; ou le tuple clé/valeur |
| if … ((not) in):  elif … : else :                   | Test/Condition                                            |
| for … in … : ;  for … in range (…) : ;  while … :   | Boucles/Iterations                                        |
| def nom_fonction() :         return                 | Définir une fonction ; Eventuellement retourner une valeur|
