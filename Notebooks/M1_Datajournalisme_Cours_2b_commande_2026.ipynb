# Récapitulatif général des commandes de base

## Numpy
|                     **Commande**                    |                         **Résultat**                   |
|:------------------------------------------------------------------------------------------------------------:|:-------------------------------------------------------------------------------------------------------------:|
| np.array                                            | Créer un tableau                                       |
| nom_variable[l,c]                                   | Accéder aux données                                    |
| nom_variable.ndim                                   | Afficher le nombre de dimensions                       |
| nom_variable.size                                   | Afficher le nombre d'éléments                          |
| nom_variable.shape                                  | Afficher la forme du tableau                           |
| np.zeros((l,c)), np ones ((l,c)), np full((l,c), v) | Créer des tableaux pré-remplis                         | 
| np.random.randn(l,c)                                |  Créer des tableaux pré-remplis avec des nombres aléatoires, mais suivant une loi normale centrée réduite (moyenne de la distribution à 0, et écart-type de 1).                                                                                                            |
| np.arange(début, fin, pas)                          | Créer un tableau à 1 dimensions allant d'un nombre de **début** à un nombre de **fin** en respectant le **pas**                                                      |
| np.mean(nom_variable)                               | Calculer la moyenne de la variable                     |
| np.median(nom_variable)                             | Calculer la médiane de la variable                     |
| np.var(nom_variable)                                | Calculer la variance de la variable                    |
| np.std(nom_variable)                                | Calculer l'écart-type de la variable                   |
| np.min(nom_variable) np.max(nom_variable)           | Afficher la valeur minimum/maximum de la variable      |
| np.sum(nom_variable)                                | Calculer la somme de la variable                       |
| nom_variable.sort()                                 | ordonne les valeurs par ordre croissant                |
| np.vstack((nom_variable, nom_variable2))            | Ajouter une ligne au tableau                           |



## Pandas

|                     **Commande**                    |                         **Résultat**                   |
|:------------------------------------------------------------------------------------------------------------:|:-------------------------------------------------------------------------------------------------------------:|
| pd.DataFrame([[0, 1, 2, 3], [4, 5, 6, 7]])          | Créer un tableau à partir de listes                    |
| pd.DataFrame(dictionnaire)                          | Créer un tableau à partir de d'un dictionnaire         |
| pd.read_csv("nom_fichier.csv")                      | Importer un jeu de données, ici au format .csv         |
| df.head()                                           | Afficher les 10 premiers éléments                      |
| df.columns                                          | Afficher le nombre de colonnes                         |
| df.index                                            | Afficher le nombre de lignes                           |
| df["nom_colonne"]                                   | Afficher les valeurs d'une colonne                     |
| df.iloc[l]                                          | Afficher le contenu d'une ligne selon l'index          |
| df.loc[1, "age"]                                    | Afficher le contenu d'une cellule selon son index et le nom d'une colonne                                                                                              |
| df.drop(["nom_colonne"], axis=1)                    | Supprimer une colonne                                  |
| df.info()                                           | Afficher l'ensemble des information du dataframe       |
| df.describe()                                       | Calculer l'ensemble des indicateurs des variables quantitatives                                                                                                  |
| df.dropna(axis=0)                                   | Supprimer les individus avec des valeurs manquantes    |
| df.nunique()                                        | Récupérer le nombre de modalités par variable          |
| df["nom_colonne"].value_counts()                    | Afficher les effectifs des modalités d'une variable    |
| df["nom_colonne"].value_counts(normalize=True).round(2)| Afficher les fréquences des modalités d'une variable                                                                                                       |
| df.groupby(["nom_colonne"])                         | Regrouper les données selon un critère                 | 
| df["nom_colonne"].unique()                          | Afficher les différentes modalités pour une variable   |
| df_YA.sort_values("nom_colonne", ascending=False)   | Ordonner les individus par ordre décroissant en fonction d'un variable                                                                                                  |
| df_YA["nom_colonne"].map(nom_fonction)              | Appliquer une fonction ou d'autres modifications sur l'ensemble d'une colonne du tableau                                                                             |
| df_YA["nom_colonne"].astype("category").cat.categories | Recoder des variables catégorielles                  |
| df.corr()                                           | Générer une matrice de corrélation                      |
| pd.crosstab(df["nom_colonne1"],df["nom_colonne2"])  | Observer les effectifs concrets dans un tableau croisé entre des variables                                                                                             |
| pd.cut(df["nom_colonne"], bins=[df["nom_colonne"].min(), n1, n2, n3, df["nom_colonne"].max()])                | Découper les valeurs d'une colonne en intervalles constants de valeur *n* ou de façon manuelle                                                                                                        | 
| pd.qcut(df["nom_colonne"], q=n)                     |Découper les valeurs d'une colonne en fonction d'un nombre *q* de classes                                                                                                  |
| df["nom_colonne"].quantile(0.25)                    | Afficher le premier quartile                            |
| df["nom_colonne"].quantile(0.75)                    | Afficher le troisième quartile                          |


## Matplotlib

|                     **Commande**                    |                         **Résultat**                   |
|:------------------------------------------------------------------------------------------------------------:|:-------------------------------------------------------------------------------------------------------------:|
| plt.figure()                                        | Créer une figure                                       |
| plt.plot(variable1, variable2)                      | Créer un graphique                                     |
| plt.show()                                          | Afficher un graphique                                  |
| plt.ylim(min, max)                                  | Limiter en ordonnée                                    |
| plt.xlim(min, max)                                  | Limiter en abscisse                                    |
| plt.ylabel("")                                      | Nom en ordonnée                                        |
| plt.xlabel("")                                      | Nom en abscisse                                        |
| plt.title("")                                       | Titre du graphique                                     |
| plt.legend()                                        | Afficher la légende                                    |
| plt.savefig("nom_figure.png")                       | Sauvegarder une figure                                 |
| plt.scatter(x,y,c=)                                 | Créer un nuage de point                                |
| df.plot(kind='bar')                                 | Créer un graphique en barre                            |
| plt.hist(df["nom_colonne"])                         | Créer un histogramme                                   |
| plt.hist2d(x,y, cmap="")                            | Créer une carte de chaleur                             |
